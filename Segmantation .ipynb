# Data Preprocessing 

```python
df.rename(columns = {'DATE' : 'Date', 'STORE_NBR' : 'Store Number', 'LYLTY_CARD_NBR':'Loyality Card Number', 'PROD_NBR':'Product Number','PROD_NAME':'Product Name', 'PROD_QTY' : 'Product Quantity', 'TOT_SALES': 'Total Sales', 'LIFESTAGE' : 'Life Stage', 'PREMIUM_CUSTOMER' : 'Premium Customer'}, inplace = True)
```

```python
df.head()
```

```python
rfm= df.groupby('Loyality Card Number').agg({'Date': lambda date: (datetime.today() - date.max()).days,
                                        'TXN_ID': lambda num: len(num),
                                        'Total Sales': lambda price: price.sum()})
```

```python
rfm.head()
```

```python
rfm.tail(80)
```

```python
rfm.columns
```

```python
rfm.columns=['monetary','frequency','recency']
```

```python
rfm['recency'] = rfm['recency'].astype(int)
rfm.head()
```

```python
rfm=rfm[['monetary','frequency','recency']].drop_duplicates()
```

```python
rfm.head()
```

```python
pd.qcut(rfm['recency'], 4)
rfm['r_quartile'] = pd.qcut(rfm['recency'].rank(method='first'), 4, ['1','2','3','4'])
```

```python
pd.qcut(rfm['monetary'], 4)
rfm['m_quartile'] = pd.qcut(rfm['monetary'].rank(method='first'), 4, ['4','3','2','1'])
```

```python
pd.qcut(rfm['frequency'], 4)
rfm['f_quartile'] = pd.qcut(rfm['frequency'].rank(method='first'), 4, ['4','3','2','1'])
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```
